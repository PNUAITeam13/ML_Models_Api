```python
from sklearn.preprocessing import OrdinalEncoder

bool_cols = ['famsup', 'paid', 'activities', 'higher', 'internet', 'romantic']
label_X_train = X_train.copy()
label_X_valid = X_valid.copy()

encoder = OrdinalEncoder()

label_X_train[bool_cols] = encoder.fit_transform(label_X_train[bool_cols])
label_X_valid[bool_cols] = encoder.transform(label_X_valid[bool_cols])
label_X_train
```

```python
X_train = label_X_train.copy()
X_valid = label_X_valid.copy()
```

```python
from sklearn.preprocessing import OneHotEncoder

OH_cols = ['reason']

OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
OH_cols_train = pd.DataFrame(OH_encoder.fit_transform(X_train[OH_cols]))
OH_cols_valid = pd.DataFrame(OH_encoder.transform(X_valid[OH_cols]))

OH_cols_train.index = X_train.index
OH_cols_valid.index = X_valid.index

num_X_train = X_train.drop(OH_cols, axis=1)
num_X_valid = X_valid.drop(OH_cols, axis=1)

OH_X_train = pd.concat([num_X_train, OH_cols_train], axis=1)
OH_X_valid = pd.concat([num_X_valid, OH_cols_valid], axis=1)

OH_X_train.columns = OH_X_train.columns.astype(str)
OH_X_valid.columns = OH_X_valid.columns.astype(str)

OH_X_train
                          
```

```python
from linear.linear_regression import LinearRegression
from sklearn.linear_model import LinearRegression as SklearnLinearRegression
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBRegressor
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_absolute_error, accuracy_score
```

```python
models = {
    'Our custom Linear Regression': LinearRegression(n_estimators=900, learning_rate=0.01),
    'sklearn Linear Regression': SklearnLinearRegression(),
    'Linear SVM Classifier model': LinearSVC(max_iter=1000, dual=False),
    'Linear Logistic Regression Classifier': LogisticRegression(max_iter=5000, C=0.1),
    'Random Forest Regressor': RandomForestRegressor(n_estimators=400, max_depth=10),
    'Random Forest Classifier': RandomForestClassifier(max_depth=10, min_samples_leaf=4, min_samples_split=10, n_estimators=200),
    'XGBoost Regressor': XGBRegressor(n_estimators=100, learning_rate=0.01, max_depth=5),
    'XGBoost Classifier': XGBClassifier(objective='multi:softmax', num_class=len(set(y_train)), n_estimators=2000, learning_rate=0.1, max_depth=5)
}
```

```python
def get_predictions(model, X_t=X_train, X_v=X_valid, y_t=y_train, y_v=y_valid):
    if isinstance(model, LinearRegression):
        model.fit(X_t, y_t, 14)
        return model.predict(X_v)
    if isinstance(model, XGBClassifier):
        label_encoder = LabelEncoder()
        y_t_encoded = label_encoder.fit_transform(y_t)
        model.fit(X_t, y_t_encoded)
        return model.predict(X_v)
    model.fit(X_t, y_t)
    return model.predict(X_v)

for name, model in models.items():
    preds = get_predictions(model)
    print(f"{name} MAE: {mean_absolute_error(y_valid, preds)}")
```

```python
# One Hot
label_X_test = X_test_full.copy()
label_X_test[bool_cols] = encoder.fit_transform(label_X_test[bool_cols])
label_X_test
```

```python
X_test = label_X_test.copy()

OH_cols_test = pd.DataFrame(OH_encoder.fit_transform(X_test[OH_cols]))
OH_cols_test.index = X_test.index
num_X_test = X_test.drop(OH_cols, axis=1)
OH_X_test = pd.concat([num_X_test, OH_cols_test], axis=1)
OH_X_test.columns = OH_X_test.columns.astype(str)
OH_X_test
```

```python
X_test_processed = OH_X_test.copy()
```

```python
regressors = {}
classifiers = {}

for name, model in models.items():
    if 'classifier' in name.lower():
        classifiers[name] = model
    else:
        regressors[name] = model
```

```python
for name, model in regressors.items():
    predictions = model.predict(X_test_processed)
    mae = mean_absolute_error(y_test, predictions)
    print(f"{name}: {mae}")

for name, model in classifiers.items():
    predictions = model.predict(X_test_processed)
    a_score = accuracy_score(y_test, predictions)
    print(f"{name}: {a_score}")
```
