```python
from data_transformer.DataTransformer import DataTransformer
transformer = DataTransformer()

data_train_X = transformer.use_ordinal_encoder(data_train_X, ["map", "bomb_planted"]).get_data()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

sse = []

k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    kmeans.fit(data_train_X)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(k_range, sse, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Sum of Squared Distances')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
k = 4
kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)

y = kmeans.fit_predict(data_train_X)

df = data_train_X.copy()
df["Cluster"] = y
```

```python
df.head(10)
```

```python
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt

reduced_data = PCA(n_components=2).fit_transform(data_train_X)
reduced_y = kmeans.fit_predict(reduced_data)

results = pd.DataFrame(reduced_data, columns=['pca1', 'pca2'])
results["Cluster"] = reduced_y

centroids = kmeans.cluster_centers_

sns.scatterplot(x="pca1", y="pca2", hue="Cluster", data=results)

plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=50, c='red', label='Centroids')

plt.title('K-means Clustering with 2 dimensions and centroids')
plt.legend()
plt.show()
```

```python
pl = sns.countplot(x='Cluster', data=df, hue=df["Cluster"])
pl.set_title("Сluster distribution graph")
plt.show()
```

```python
large_features = ["t_money", "ct_money", "ct_score", "t_score","t_armor", "ct_armor", "t_health", "ct_health"]
small_features = [feature for feature in features if feature not in large_features]
```

```python
melted_data = df.melt(id_vars='Cluster', value_vars=large_features, var_name='Feature', value_name='Value')

for feature in large_features:
    plt.figure(figsize=(6, 3))
    for cluster in melted_data['Cluster'].unique():
        cluster_values = melted_data[(melted_data['Feature'] == feature) & (melted_data['Cluster'] == cluster)]['Value']
        sns.histplot(cluster_values, label=f'Cluster {cluster}', alpha=0.5)
    
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.title(f'Distribution of {feature} by Cluster')
    plt.legend()
    plt.show()
```

```python
plots_per_row = 3
chunks = [small_features[i:i + plots_per_row] for i in range(0, len(small_features), plots_per_row)]

for chunk in chunks:
    fig, axes = plt.subplots(1, len(chunk), figsize=(15, 4), sharey=True)

    for ax, feature in zip(axes, chunk):
        sns.countplot(data=df, x=feature, hue='Cluster', palette='viridis', ax=ax)
        ax.set_title(f'Cluster-wise Count of {feature}')
        ax.legend(title='Cluster')
        ax.tick_params(axis='x', rotation=45)

    plt.tight_layout()
    plt.show()
```

```python
# And here fitting on model is started...
from sklearn.model_selection import cross_val_score, train_test_split
from xgboost import XGBRegressor, XGBClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, accuracy_score, mean_squared_error, f1_score
import numpy as np
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, data_train_y, train_size=0.8, test_size=0.2)
```

```python
def run_simply_regressor_model(X_train, y_train, X_valid, y_valid, model, estimators = [100]):
    best_model = None
    best_score = None
    for n_estimator in estimators:
        regressor = model(n_estimators=n_estimator, max_depth=10)
        regressor.fit(X_train, y_train)
        predictions = regressor.predict(X_valid)
        
        mse = mean_squared_error(y_valid, predictions)
        mae = mean_absolute_error(y_valid, predictions)
        print(f"{model.__name__} ({n_estimator}): MSE - {mse}; MAE - {mae}")
        if best_model is None or mae < best_score:
            best_model = regressor
            best_score = mae
    
    return best_score
```

```python
def run_xgboost_regressor(X_train, y_train, X_valid, y_valid, estimators=[800]):
    best_model = None
    best_score = None
    for n_estimator in estimators:
        xgboost_regressor = XGBRegressor(n_estimators=n_estimator, learning_rate=0.04, max_depth=9)
        xgboost_regressor.fit(X_train, y_train)
        predictions = xgboost_regressor.predict(X_valid)
        mse = mean_squared_error(y_valid, predictions)
        mae = mean_absolute_error(y_valid, predictions)
        print(f"XGBOOST Regressor ({n_estimator}): MSE - {mse}; MAE - {mae}")

        if best_model is None or mae < best_score:
            best_model = xgboost_regressor
            best_score = mae

    return best_model
```

```python
def run_xgboost_classifier(X_train, y_train, X_valid, y_valid, estimators=[1000]):
    best_model = None
    best_score = None
    for n_estimator in estimators:
        xgboost_classifier = XGBClassifier(objective='multi:softmax', num_class=len(set(y_train)), n_estimators=n_estimator, 
                                           learning_rate=0.1, max_depth=11)
        xgboost_classifier.fit(X_train, y_train)
        predictions = xgboost_classifier.predict(X_valid)
        a_score = accuracy_score(y_valid, predictions)
        print(f"XGBOOST Classifier: Accuracy Score - {a_score}, F1 Score - {f1_score(y_valid, predictions)}")
        if best_model is None or a_score > best_score:
            best_model = xgboost_classifier
            best_score = a_score

    return best_model
```
